```python
def report3plus(string):
    if string.count("'") > 2:
        print(string)
```

```python
def check3plus(lst):
    for string in lst:
        report3plus(string)
```

```python
def check_ing(lst):
    for string in lst:
        if len(string) > 3:
            if string[0] != "'" and string[-3:] == "in'":
                print(string)
```

```python
def check_double_quotes(lst):
    for string in lst:
        if string.count("'") == 2:
            if (string[0] == "'" and string[-1] != "'") or (string[0] != "'" and string[-1] == "'"):
                print(string)
```
