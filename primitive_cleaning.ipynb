```python
import numpy as np
import pandas as pd
import re
import string
```

```python
def keep_printable(text):
    return ''.join(list(filter(lambda x: x in string.printable, text)))
```

```python
def scan_text(series):
    return series.apply(keep_printable)
```

```python
def check_length(series):
    return (series.str.len() < 320).value_counts()
```

```python
def keep_songs(series, length=320):
    return series.loc[series.str.len() > length]
```

```python
songs = pd.concat([
    allsongs.lyrics,
    chance.Lyrics,
    emi.text,
    fromweb.Lyrics,
    kanye.lyric,
    lyrics.lyrics,
    lyrics1.Lyrics,
    lyrics2.Lyrics,
    songdat.text,
    steve.lyrics
])
```

```python
(songs.dropna().str.len() < 320).value_counts()
```

```python
cleaned_songs = scan_text(songs.dropna())
```

```python
(cleaned_songs.str.len() < 320).value_counts()
```

```python
keep_songs(cleaned_songs).to_csv('lyrics_files/cleaned_lyrics', index=False, header=None)
```
